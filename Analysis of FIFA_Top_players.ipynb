# columns need to be converted for further analysis:

# "value" - Convert to Float after getting rid of currency symbol and suffix.

# 'Wage'- - Convert to Float after getting rid of currency symbol and suffix.

```python
df["Value"]
```

```python
def myfunction(column):
    if isinstance(column,str):
        a=column.replace("€","")
        
        if "M" in a:
            a=float(a.replace("M",""))*1000000
        elif "K" in a:
            a=float(a.replace("K",""))*1000
        return float(a)
```

```python
df["Value"]=df["Value"].apply(myfunction)
df["Wage"]=df["Wage"].apply(myfunction)
```

```python
df["Wage"]
```

# 'Height' -Convert to Float with decimal points

```python
# Height is in feet and inch convert to feet.
# 5 feet 7 inch = ((5*12)+7)/12 feet
```

```python
def myfunction1(column):
    l=[]
    s=1
    
    if isinstance(column,str):
        l=column.split("'")
        i=int(l[0])
        j=int(l[1])
        s=((i*12)+j)/12
    return(s)
         
```

```python
df["Height"]=df["Height"].apply(myfunction1)
```
